```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.__version__
```

```python
import warnings
warnings.filterwarnings("ignore")
```

**1. Style and color**

```python
sns.set_theme(font= "serif")
```

**2. Read csv file**

```python
data = pd.read_csv("StudentsPerformance.csv")
data.head()
```

```python
sns.displot(data= data, 
            x= "math score", 
            y= "reading score", 
            kind= "kde", 
            rug= True, 
            hue= "gender", 
            col= "race/ethnicity", 
            row= "parental level of education", 
            palette="Accent")
```

```python
sns.barplot(data= data, x= "gender", y= "math score")
```

```python
data.head()
```

```python
sns.barplot(data= data, x= "race/ethnicity", y= "math score", hue= "gender", ci= 95,estimator= np.std)
```

```python
data.head()
```

```python
sns.boxplot(data= data, x= "math score", y= "gender")
```

```python
sns.stripplot(data= data, x= "math score", y= "gender")
sns.boxplot(data= data, x= "math score", y= "gender")
```

```python
sns.violinplot(data= data, x= "math score", y= "gender")
```

```python
sns.lmplot(data= data, x= "math score", y= "writing score", hue= "gender")
```

```python
data.head()
```

```python
sns.jointplot(data= data, x= "math score", y= "reading score", kind= "kde", hue= "gender")
```

```python
sns.pairplot(data, kind= "kde", diag_kind= "hist")
```

```python
data.head()
```

```python
g = sns.FacetGrid(data, col= "gender", row= "lunch", size= 5)
g.map(plt.hist, 'math score')
```

**Heatmap**

```python
data.head()
```

```python
corr = data.corr()
matrix = np.triu(corr)
matrix
```

```python
sns.heatmap(corr, vmin= 0, vmax= 1, annot=True, mask= matrix)
```
